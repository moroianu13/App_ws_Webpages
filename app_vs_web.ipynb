## 1. Trend Forecasting with Linear Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Create the DataFrame
data = {
    'Year': [2020, 2021, 2022, 2023],
    'Web Visits': [100000, 110000, 115000, 120000],
    'App Visits': [60000, 80000, 100000, 140000],
    'Unique Mobile Visitors': [10000, 12000, 15000, 20000]
}

df = pd.DataFrame(data)

# Prepare data for regression
X = df[['Year']].values
y_web = df['Web Visits'].values
y_app = df['App Visits'].values

# Create and fit the model
model_web = LinearRegression()
model_web.fit(X, y_web)
model_app = LinearRegression()
model_app.fit(X, y_app)

# Predict future values
future_years = np.array([[2024], [2025], [2026]])
pred_web = model_web.predict(future_years)
pred_app = model_app.predict(future_years)

# Plot historical and forecasted values
plt.figure(figsize=(14, 7))
plt.plot(df['Year'], df['Web Visits'], marker='o', label='Web Visits (Historical)')
plt.plot(df['Year'], df['App Visits'], marker='o', label='App Visits (Historical)')
plt.plot(future_years, pred_web, linestyle='--', color='blue', label='Web Visits (Forecast)')
plt.plot(future_years, pred_app, linestyle='--', color='orange', label='App Visits (Forecast)')
plt.xlabel('Year')
plt.ylabel('Visits')
plt.title('Web vs. App Visits Forecast')
plt.legend()
plt.grid(True)
plt.show()
```

## 2. Time Series Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Sample data
data = {
    'Year': ['2020', '2021', '2022', '2023'],
    'Web Visits': [100000, 110000, 115000, 120000],
    'App Visits': [60000, 80000, 100000, 140000],
}

df = pd.DataFrame(data)
df['Year'] = pd.to_datetime(df['Year'], format='%Y')
df.set_index('Year', inplace=True)

# Convert to time series
web_series = df['Web Visits']
app_series = df['App Visits']

# Decompose time series
decomp_web = sm.tsa.seasonal_decompose(web_series, model='additive', period=1)
decomp_app = sm.tsa.seasonal_decompose(app_series, model='additive', period=1)

# Plot decomposed components
plt.figure(figsize=(14, 7))

plt.subplot(2, 2, 1)
decomp_web.trend.plot()
plt.title('Web Visits Trend')

plt.subplot(2, 2, 2)
decomp_web.seasonal.plot()
plt.title('Web Visits Seasonality')

plt.subplot(2, 2, 3)
decomp_web.resid.plot()
plt.title('Web Visits Residuals')

plt.subplot(2, 2, 4)
decomp_app.trend.plot()
plt.title('App Visits Trend')

plt.tight_layout()
plt.show()
```

## 3. User Segmentation Analysis

```python
from sklearn.cluster import KMeans
import os
os.environ['LOKY_MAX_CPU_COUNT'] = '4'

features = {
    'Web Visits': [100000, 110000, 115000, 120000],
    'App Visits': [60000, 80000, 100000, 140000]
}
df_features = pd.DataFrame(features)

# Apply K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=0).fit(df_features)
df_features['Cluster'] = kmeans.labels_

# Plot clusters
plt.figure(figsize=(10, 6))
plt.scatter(df_features['Web Visits'], df_features['App Visits'], c=df_features['Cluster'], cmap='viridis')
plt.xlabel('Web Visits')
plt.ylabel('App Visits')
plt.title('User Segmentation with K-Means Clustering')
plt.colorbar(label='Cluster')
plt.show()
```
